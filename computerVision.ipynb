```python
import cv2
import numpy as np
import requests
from matplotlib import pyplot as plt

# URL of the image (make sure it's the direct download URL for Google Drive)
url = 'https://drive.google.com/uc?id=1Ur-c8fpvOlgKfeZWZYDo1yhVT1_mCVNt'  # Replace with your file's ID

# Get the image from the URL
response = requests.get(url)
image_array = np.asarray(bytearray(response.content), dtype=np.uint8)

# Decode the image into an OpenCV format
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

# Check if image was loaded successfully
if image is not None:
    print("Image loaded successfully")

    # Convert BGR to RGB (OpenCV loads images in BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.imshow(image_rgb)
    plt.title('Loaded Image from URL')
    plt.axis('off')  # Hide axes
    plt.show()

else:
    print("Failed to load image")
```

```python

# Convert BGR to RGB for displaying using matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(image_rgb)
plt.title('Original Image')
plt.show()
```

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

```python
# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
image_contours = image.copy()
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)

# Convert BGR to RGB for displaying using matplotlib
image_contours_rgb = cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB)

# Display the image with contours
plt.imshow(image_contours_rgb)
plt.title('Contours Image')
plt.show()
```
